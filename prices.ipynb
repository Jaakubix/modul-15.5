```python
import pandas as pd
import matplotlib.pyplot as plt


prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25)
]
months = []
price_values = []

for month in prices:
    months.append(month[0])

for price in prices:
    price_values.append(price[1]) 

df = pd.DataFrame(prices)
df.columns = [ 'month' , 'price']
df
```

```python
plt.title(' Prices of goods [usd]')
plt.xlabel('months')
plt.ylabel('prices [usd]')
plt.plot(months,price_values, 'r--')
```
